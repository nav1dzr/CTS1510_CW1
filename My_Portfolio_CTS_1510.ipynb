```python
## the scientific converter including meanly lambda functions.

def length():

    print("1. m to ft")
    print("2. in to cm")
    print("3. mi to km")

    choice = input("Enter your choice: ")
    value = float(input("Enter value: "))

    m_to_ft = lambda x: x * 3.28
    in_to_cm = lambda x: x * 2.54
    mi_to_km = lambda x: x * 1.61

    if choice == "1":
        print("Result:", m_to_ft(value))
    elif choice == "2":
        print("Result:", in_to_cm(value))
    elif choice == "3":
        print("Result:", mi_to_km(value))
    else:
        print("Invalid choice!")


def temperature():
    print("1. °C to °F")
    print("2. °F to °C")

    choice = input("Enter your choice: ")
    value = float(input("Enter value: "))

    c_to_f = lambda x: (x * 9/5) + 32
    f_to_c = lambda x: (x - 32) * 5/9

    if choice == "1":
        print("Result:", c_to_f(value))
    elif choice == "2":
        print("Result:", f_to_c(value))
    else:
        print("Invalid choice!")


print("Choose a category:")
print("1. Length")
print("2. Temperature")

main_choice = input("Enter your choice: ")

if main_choice == "1":
    length()
elif main_choice == "2":
    temperature()
else:
    print("Invalid category!")
```

```python
## Continue on My program that's takes name and bank details to register for membership
import datetime

show_time = lambda: print(datetime.datetime.now().strftime("\n%Y-%m-%d %H:%M:%S"))


check_bank = lambda x: x.isdigit() and len(x) == 8
check_sort = lambda x: x.isdigit() and len(x) == 6

print("=== Gym Membership Registration ===\n")

name = input("what is your name? ")


while True:
    bank = input("Enter your bank account number (8 digits): ")
    if check_bank(bank):
        break
    else:
        print("Bank account number must be 8 digits!")

while True:
    sortcode = input("Enter your sort code (6 digits): ")
    if check_sort(sortcode):
        break
    else:
        print("❌ Sort code must be 6 digits!")


memberships = [
    "(1) Daily  (£9.99)",
    "(2) Weekly (£69.99)",
    "(3) Monthly (£199.99)"
]

print("\nAvailable Memberships:")
for plan in memberships:
    print(plan)

while True:
    member = input("\nChoose your membership (1, 2, or 3): ")

    if member.isdigit():  # check if it's a number
        member = int(member)
        if 1 <= member <= 3:
            chosen = memberships[member - 1]
            break
        else:
            print(" Please choose a valid option (1–3).")

# --- Confirmation ---
print("\n Registration Successful!")
print("-----------------------------")
print(f"Name: {name}")
print(f"Bank Account: {bank}")
print(f"Sort Code: {sortcode}")
print(f"Membership: {chosen}")

# --- Print time ---
show_time()
```
